```python
import pandas as pd

data = pd.read_csv('Bangalore_AQI_Dataset.csv')
# Fix date parsing by specifying dayfirst=True
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Check missing values
print(data.isnull().sum())

# Impute missing pollutant values with median
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']
for col in pollutants:
    data[col] = data[col].fillna(data[col].median())


# Drop rows with missing AQI if any
data.dropna(subset=['AQI'], inplace=True)
```

```python
print(data.isnull().sum())
```

```python
# Extract date features
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.weekday

# Normalize pollutant features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[pollutants] = scaler.fit_transform(data[pollutants])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(x='Date', y='PM2.5', data=data)
plt.title('PM2.5 over time')
plt.savefig('pm25_over_time.png')
# plt.show()  # omit this line
```
